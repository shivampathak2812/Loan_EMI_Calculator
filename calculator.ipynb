```python
import math

def calculate_emi(amount, rate, years):
    r = rate / (12 * 100)
    n = years * 12
    emi = amount * r * ((1 + r) ** n) / ((1 + r) ** n - 1)
    return emi

def generate_schedule(amount, rate, years):
    r = rate / (12 * 100)
    n = years * 12
    emi = calculate_emi(amount, rate, years)

    balance = amount
    schedule = []

    for i in range(1, n + 1):
        interest = balance * r
        principal = emi - interest
        balance -= principal
        schedule.append([i, round(principal, 2), round(interest, 2), round(balance, 2)])

    return schedule
```
