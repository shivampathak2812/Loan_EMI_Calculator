```python
import matplotlib.pyplot as plt
```

```python
def plot_emi_chart(schedule):
    months = [row["Month"] for row in schedule]
    interest = [row["Interest Paid"] for row in schedule]
    principal = [row["Principal Paid"] for row in schedule]

    plt.figure(figsize=(11,5))
    plt.plot(months, interest, label="Interest Portion")
    plt.plot(months, principal, label="Principal Portion")

    plt.xlabel("Month")
    plt.ylabel("Amount (â‚¹)")
    plt.title("EMI Interest vs Principal Breakdown")
    plt.legend()
    plt.grid(True)
    plt.show()
```
